```python
import yfinance as yf
import pandas as pd
# Download code
# Set up the ticker
ticker = yf.Ticker("^SPX")  # You can replace this with another symbol if needed
expirations = ticker.options

# Initialize a list to store DataFrames
all_calls = []

# Loop through each expiration date
for date in expirations:
    opt = ticker.option_chain(date)
    calls = opt.calls.copy()
    calls['expirationDate'] = date  # Add expiration date column
    all_calls.append(calls)

# Combine all call options into a single DataFrame
calls_df = pd.concat(all_calls, ignore_index=True)

# Save to CSV
calls_df.to_csv("2025-08-13spx_call_options.csv", index=False)

print("CSV file 'spx_call_options.csv' created with all call option data.")
```
